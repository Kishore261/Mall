```python
q1=df['Annual Salary'].quantile(0.25)
q3=df['Annual Salary'].quantile(0.75)
IQR=q3-q1
lb=q1-(1.5*IQR)
ub=q3+(1.5*IQR)
print(lb)
print(ub)
```

```python
median_val=df['Annual Salary'].median()
print(median_val)
```

```python
df['Annual Salary']=np.where(df['Annual Salary']>ub,median_val,df['Annual Salary'])
```

```python
df['Annual Salary']=np.where(df['Annual Salary']<lb,median_val,df['Annual Salary'])
```
